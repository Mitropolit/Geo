```python
import pandas as pd
import numpy as np
import rasterio
import geopandas
from shapely.geometry import Point, mapping
import json
```

```python
clay_path = rasterio.open('clay.tif')
sand_path = rasterio.open('sand.tif')
density_path = rasterio.open('density.tif')
moisture_path = rasterio.open('soil_moisture.tif')
fields = geopandas.read_file("field_centroids.geojson")
```

```python
field_list = []

# –í–∏—Ç—è–≥—É—î–º–æ –¥–∞–Ω—ñ –ø—Ä–æ –≥—Ä—É–Ω—Ç —Ç–∞ –≤–æ–ª–æ–≥—ñ—Å—Ç—å
for index, row in fields.iterrows():
    # –ù–∞–∑–≤–∞ –ø–æ–ª—è —Ç–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ —Ü–µ–Ω—Ç—Ä–æ—ó–¥–∞
    field_id = row['id']
    field_name = row['Name']
    centroid = row['geometry'].centroid
    lat, lon = centroid.y, centroid.x

    # –ü—ñ–∫—Å–µ–ª—å–Ω—ñ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ —Ü–µ–Ω—Ç—Ä–æ—ó–¥–∞
    row, col = clay_path.index(lon, lat)

    # –î–∞–Ω—ñ –ø—Ä–æ “ë—Ä—É–Ω—Ç –¥–ª—è —Ü–µ–Ω—Ç—Ä–æ—ó–¥–∞
    clay = clay_path.read(1)[row, col]
    sand = sand_path.read(1)[row, col]
    density = density_path.read(1)[row, col]

    # –ü—ñ–∫—Å–µ–ª—å–Ω—ñ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ —Ü–µ–Ω—Ç—Ä–æ—ó–¥–∞ –¥–ª—è –¥–∞–Ω–∏—Ö –ø—Ä–æ –≤–æ–ª–æ–≥—ñ—Å—Ç—å
    row, col = sand_path.index(lon, lat)

    # –î–∞–Ω—ñ –ø—Ä–æ –≤–æ–ª–æ–≥—ñ—Å—Ç—å
    moisture = moisture_path.read(1)[row, col]

    # –°–ª–æ–≤–Ω–∏–∫ —ñ–∑ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—î—é –ø—Ä–æ –ø–æ–ª—è
    field_dict = {
        'type': 'Feature',
        'properties': {
            'id': field_id,
            'name': field_name,
            'soil': {
                'clay': float(clay),
                'sand': float(sand),
                'density': int(density)
            },
            'soil_moisture': int(moisture),
            'coordinates': {
                'lat': float(lat),
                'lng': float(lon)
            }
        },
        'geometry': mapping(centroid)
    }

    # –°–ª–æ–≤–Ω–∏–∫ –ø–æ–ª—è –¥–æ —Å–ø–∏—Å–∫—É
    field_list.append(field_dict)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É –ø–æ–ª—ñ–≤ –Ω–∞ GeoJSON FeatureCollection
feature_collection = {
    'type': 'FeatureCollection',
    'features': field_list
}

# –ó–∞–ø–∏—Å GeoJSON —É —Ñ–∞–π–ª
with open('field_data.geojson', 'w') as f:
    json.dump(feature_collection, f, indent=2)
   
```
